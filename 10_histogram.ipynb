```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import pandas as pd
```

```python
path = 'images/share.jpeg'
```

```python
img = cv2.imread(path)
blue = cv2.calcHist([img],[0],None, [256],[0,256])
green = cv2.calcHist([img],[1],None, [256],[0,256])
red = cv2.calcHist([img],[2],None, [256],[0,256])
plt.plot(blue,color = "b")
plt.plot(green,color = "green")
plt.plot(red,color = "red")
plt.title('histogram')
plt.xlabel('pixel value')
plt.ylabel('freq')
plt.show()
```

```python

img = cv2.imread(path)
name = 'trackbar'
cv2.namedWindow(name)
cv2.createTrackbar('Threshold1',name,0,255, lambda x : x)
cv2.createTrackbar('Threshold2',name,0,255, lambda x : x)

while True:
    threshold1 = cv2.getTrackbarPos('threshold1',name)
    threshold2 = cv2.getTrackbarPos("threshold2",name)
    canny_frame = cv2.Canny(img,threshold1,threshold2)

    cv2.imshow(name,canny_frame)
    if cv2.waitKey(1) == ord("q"):
        break

cv2.destroyAllWindows()

cv2.waitKey(1)
```

```python
cap = cv2.VideoCapture(0)
name = 'trackbar'
cv2.namedWindow(name)
cv2.createTrackbar('Threshold1',name,0,255, lambda x : x)
cv2.createTrackbar('Threshold2',name,0,255, lambda x : x)

if not cap.isOpened():
    exit()
while True:
    threshold1 = cv2.getTrackbarPos('threshold1',name)
    threshold2 = cv2.getTrackbarPos("threshold2",name)
    ret, frame = cap.read()
    canny_frame = cv2.Canny(frame,threshold1,threshold2)
    if not ret
        break
    cv2.imshow(name,canny_frame)
    if cv2.waitKey(1) == ord("q"):
        break
cap.release()

cv2.destroyAllWindows()

cv2.waitKey(1)
```
